```python
import sdf_helper
import numpy as np
from cmcrameri import cm as cmaps
import matplotlib, os, re, glob
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.gridspec as gridspec
import pandas as pd
from skimage.measure import block_reduce
import time
import timelord as tl

SimulationPath = "/Volumes/NKehoe/CoherentAcc/EPOCH/3PW/BayOpt/Iter/11"
```

```python
LenSim = len([int(i.split('/')[-1].split('.')[0]) for i in glob.glob(f'{SimulationPath}/*.sdf')])
Files = [sdf_helper.getdata(os.path.join(SimulationPath, f"{i:04d}.sdf"), verbose=False) for i in range(LenSim)]
```

```python
Axis = ["x", "y", "z", "Time"]
for ax in Axis:
    print(f"Axis: {ax}")
    print(Axis)
    Axis.remove(ax)  # Remove axis from AxisNames to avoid duplication
    print(Axis)
```

```python
reduce = True
dx, dy = 4, 4
AxisNames = ['x', 'y', 'Time']
Axis = {axis: np.array(()) for axis in AxisNames}
attr = "Derived_Number_Density_electron"
Data = []
time1 = time.time()

for i, File in enumerate(Files):
    print(f"\nProcessing File {i}")
    start_time = time1
    Den = np.array(getattr(File, attr).data)
    if reduce:
        nx = Den.shape[0] // dx
        ny = Den.shape[1] // dy
        Den = np.where(Den == 0, np.nan, Den)
        with np.errstate(invalid='ignore'):
            Den = block_reduce((Den[:nx * dx, :ny * dy]), (dx, dy), np.nanmean)
    Data.append(Den)
    print(f"Data shape for file {i}: {Den.shape}")
    print(f"Data shape: {len(Data)}")
    time1 = time.time()
    print(f"File {i} processed in {time1 - start_time:.2f} seconds")

Data = np.array(Data)
time2 = time.time()
print(f"Total processing time: {time2 - time1:.2f} seconds")
print(f"Final Data shape: {Data.shape}")
```
